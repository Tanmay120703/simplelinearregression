```python
import pandas as pd
```

```python
df = pd.read_csv('//content//sample_data//cgpa_lpa_data.csv')
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(df['CGPA'] , df['LPA'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , X_test , y_train , y_test = train_test_split(df[['CGPA']] , df['LPA'] , test_size=0.2 , random_state=2)
```

```python
x_train
```

```python
lr.fit(x_train , y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
acc = mean_squared_error(y_test , y_pred)
```

```python
acc
```

```python
acc1 = mean_absolute_error(y_test , y_pred)
```

```python
acc1
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['CGPA'],df['LPA'], color='blue', label='Data Points')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Linear Regression Line')
plt.title('Linear Regression: CGPA vs LPA')
plt.xlabel('CGPA')
plt.ylabel('LPA (Lakhs Per Annum)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
